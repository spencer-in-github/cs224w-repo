```python
import numpy as np

A = np.array([[0, 1/2, 1/3, 0],
              [1/2, 0, 1/3, 0],
              [1/2, 1/2, 0, 1],
              [0, 0, 1/3, 0]])

eigvals, eigvecs = np.linalg.eig(A)

print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

```python
eigvecs[0]
```

```python
v_norm = eigvecs[0]/np.linalg.norm(eigvecs[0])
v_rouned = [round(elem, 4) for elem in v_norm]
print(v_rouned)
```
